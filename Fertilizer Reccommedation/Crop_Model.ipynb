```python
import numpy as np
import pandas as pd
import sklearn

print("Environment Ready")
```

```python
df = pd.read_csv("crop_fertilizer.csv")

df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=["Link"])
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_district = LabelEncoder()
le_soil = LabelEncoder()
le_crop = LabelEncoder()
le_fert = LabelEncoder()

df["District_Name"] = le_district.fit_transform(df["District_Name"])
df["Soil_color"] = le_soil.fit_transform(df["Soil_color"])
df["Crop"] = le_crop.fit_transform(df["Crop"])
df["Fertilizer"] = le_fert.fit_transform(df["Fertilizer"])

# Machine learning models only understand numbers.
# Kolhapur → number
# Black → number
# Sugarcane → number
# Urea → number
```

```python
df.head()  #yes All is Number
```

```python
X = df.drop(["Crop","Fertilizer"], axis=1)
```

```python
print(X) #Separate Input....
```

```python
y_crop = df["Crop"]
y_fert = df["Fertilizer"]
# Separate Input and Output
```

```python
print(X.shape)
print(y_crop.shape)
print(y_fert.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_crop_train, y_crop_test = train_test_split(
    X, y_crop, test_size=0.2, random_state=42
)

X_train2, X_test2, y_fert_train, y_fert_test = train_test_split(
    X, y_fert, test_size=0.2, random_state=42
)

#Train-Test Split
```

```python
# Step-8: Train Model
# Use Random Forest (very good for agriculture datasets).
from sklearn.ensemble import RandomForestClassifier

crop_model = RandomForestClassifier(n_estimators=100)

crop_model.fit(X_train, y_crop_train)
```

```python
fert_model = RandomForestClassifier(n_estimators=100)

fert_model.fit(X_train2, y_fert_train)
#Fertilizer model:
```

```python
#Check Accuracy  - Crop prediction:
from sklearn.metrics import accuracy_score

crop_pred = crop_model.predict(X_test)

print("Crop Accuracy:", accuracy_score(y_crop_test, crop_pred))
```

```python
#Fertilizer prediction:
fert_pred = fert_model.predict(X_test2)

print("Fertilizer Accuracy:", accuracy_score(y_fert_test, fert_pred))
```

```python
import joblib

joblib.dump(crop_model,"crop_model.pkl")
joblib.dump(fert_model,"fert_model.pkl")
```
